```python
# load libraries
%matplotlib inline
import pandas as pd
import numpy as np
import time
import os
```

```python
#read in the practice CSV as a Pandas DataFrame
df = pd.read_csv("draft8_full.csv", header=0)
#show the first five lines to verify upload
df.head()
```

```python
df = df.drop(['CourseDescription','CourseName'], axis=1)
df.head()
```

```python
df['StudentID'] = pd.Series(0,index=df.index)
df.head()
```

```python
new_cols = []
cols = df.columns.tolist()
new_cols.append(cols[0])
new_cols.append(cols[-1])
new_cols = new_cols + cols[1:-1]
df = df[new_cols]
df.head()
```

```python
for i in df.index:
    if (df.at[i,'FirstName'] == 'JULIET') and (df.at[i,'LastName'] == 'WANYIRI'):
        df.at[i,'Gender'] = 'F'
```

```python
student_list = []
new_id = 1

for i in df.index:
    first = df.at[i, 'FirstName']
    last = df.at[i, 'LastName']
    
    #check if we've already done this student
    already_done = False
    for student in student_list:
        if (student[0] == first) and (student[1] == last):
            existing_id = student[2]
            already_done = True
            break
    #if we've already set this student's ID, fill the table with the relevant one
    if already_done:
        df.at[i,'StudentID'] = existing_id
    #otherwise, put the student in the tracker with the latest ID, fill the table, and increment the ID
    else:
        student_list.append((first, last, new_id))
        df.at[i,'StudentID'] = new_id
        new_id += 1
        
df.head()
```

```python
df = df.drop(['FirstName','LastName','YearTaken'], axis=1)
df.head()
```

```python
#Save it as a CSV to the local directory
df.to_csv(r'draft8_reduced.csv', index = False)
```
