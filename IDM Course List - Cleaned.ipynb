```python
# load libraries
%matplotlib inline
import pandas as pd
import numpy as np
import time
import os
```

```python
#read in the original file from Tony as a Pandas DataFrame
df = pd.read_csv("IDM_courses_original.csv", header=0)
#show the first five lines to verify upload
df.head()
```

```python
#Initialize the output dataframe with a dummy row
init_data = [['first','last', 'year', 'course number', 'course name', 'year']]
output_df = pd.DataFrame(init_data, columns = ['First_Name','Last_name','Mat_Year', 'Course_Num', 'Course_Name', 'Year_Taken'])
#Verify the new dataframe looks as intended
output_df.head()
```

```python
#Stupid way to track the correct index of the output DF, but I can't get .append to work
k=0

#iterate through every row in the original file
for i in df.index:
    #pull out the student name and course name into temporary variables
    first = df.at[i,'FIRST_NAME']
    last = df.at[i,'LAST_NAME']
    mat_year = df.at[i,'MATRICULATION_YEAR']
    course_num = df.at[i,'COURSE']
    course_name = df.at[i,'COURSE_NAME']
    yr_taken = df.at[i,'YEAR']
        
    if (('Spec' not in course_name) and ('Studio' not in course_name) and ('Indep' not in course_name) and ('Topic' not in course_name) and ('Sem' not in course_name) and ('SIP' not in course_name) and ('EM' not in course_num) and ('ESD' not in course_num)):
        temp_data = [[first, last, mat_year, course_num, course_name, yr_taken]]
        temp_df = pd.DataFrame(temp_data, columns = ['First_Name','Last_name','Mat_Year', 'Course_Num', 'Course_Name', 'Year_Taken'])
        output_df.loc[k] = temp_df.loc[0]
        #Move to the next row in the output DF
        k += 1
        
#show the entire output df to see length
output_df
```
